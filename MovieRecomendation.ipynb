```python
# this is a string of  list,,we need to convert it into list

import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
  L =[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
# do the same thing

movies['keywords']=movies['keywords'].apply(convert)
```

```python
#cast

movies["cast"][0]
```

```python
# cast column we need first 3 actors

def convert3(obj):
  L =[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter +=1
    else:
      break

  return L

```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python


movies['crew'][0]
```

```python
# only we need director

def director(obj):
  l =[]
  for i in ast.literal_eval(obj):
    if i['job']=="Director":
      l.append(i["name"])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies["overview"]
```

```python
# converting into list

movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# replacing the space between

movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])   # replacing space(" ") with nothing ("")
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])   # replacing space(" ") with nothing ("")
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])   # replacing space(" ") with nothing ("")
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])   # replacing space(" ") with nothing ("")
```

```python
movies.head()
```

```python
# we converging few columns

movies["tags"] =movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')     #most repeated words 500 columns aakum then each sentence row aakum ennit athinte count value aakum suppose "high" is column then 1st row il 1st sentence aan athil 5 thavana repeat cheythittundenkil value will be 5;
```

```python
vec = cv.fit_transform(new_df.tags).toarray()  #out put is matrcs so convert into array
```

```python
vec
```

```python
#cv.get_feature_names()  # feature names
```

```python
#stemming

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# tags il thanne words stemming cheyyunnu

def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)   #convetinng list in to string
```

```python
# example
stem("hello marines this is your duty")
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vec)
```

```python
# sorted(similarity[0],reverse=True)  
# when we sort we lose index so we need enumarate function
# list(enumerate(similarity[0]))  #list of tuples so we get index with distance
# sorted(list(enumerate(similarity[0])),reverse=True) # sorted with index value we need to sort with second value
# sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] #we get similar 5 movies  we dont need same movie so 1 to 6
```

```python
# find 5 movies that are similar distance

def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distance=similarity[movie_index]
  movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movie_list:
    print(new_df.iloc[i[0]].title)
```
